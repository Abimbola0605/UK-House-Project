## Missing Value

```python
# Clean column names
house.columns = [col.lower().replace(' ', '_').replace('(', '').replace(')', '') for col in house.columns]
numeric_cols = house.select_dtypes(include=['float64', 'int64']).columns
house[numeric_cols] = house[numeric_cols].fillna(house[numeric_cols].mean())
```

```python
# Convert date to datetime
house['pivotable_date'] = pd.to_datetime(house['pivotable_date'])
```

```python
house['average_price_first-time_buyers'] = house['average_price_first-time_buyers'].fillna(house['average_price_all_property_types'])
house['average_price_former_owner-occupiers'] = house['average_price_former_owner-occupiers'].fillna(house['average_price_all_property_types'])
house['average_price_cash_purchases'] = house['average_price_cash_purchases'].fillna(house['average_price_all_property_types'])
house['average_price_mortgage_purchases'] = house['average_price_mortgage_purchases'].fillna(house['average_price_all_property_types'])
```

```python
house['house_price_index_first-time_buyers'] = house['house_price_index_first-time_buyers'].fillna(house['house_price_index_all_property_types'])
house['house_price_index_former_owner-occupiers'] = house['house_price_index_former_owner-occupiers'].fillna(house['house_price_index_all_property_types'])
house['house_price_index_cash_purchases'] = house['house_price_index_cash_purchases'].fillna(house['house_price_index_all_property_types'])
house['house_price_index_mortgage_purchases'] = house['house_price_index_mortgage_purchases'].fillna(house['house_price_index_all_property_types'])
```

```python
house['sales_volume_cash_purchases'] = house['sales_volume_cash_purchases'].fillna(house['sales_volume_all_property_types'])
house['sales_volume_mortgage_purchases'] = house['sales_volume_mortgage_purchases'].fillna(house['sales_volume_all_property_types'])
```

```python
percentage_change_cols = [
    'percentage_change_yearly_first-time_buyers',
    'percentage_change_monthly_first-time_buyers',
    'percentage_change_yearly_former_owner-occupiers',
    'percentage_change_monthly_former_owner-occupiers',
    'percentage_change_yearly_cash_purchases',
    'percentage_change_monthly_cash_purchases',
    'percentage_change_yearly_mortgage_purchases',
    'percentage_change_monthly_mortgage_purchases'
]
for col in percentage_change_cols:
    house[col] = house[col].fillna(0)
```

```python
house.fillna(method='ffill', inplace=True)
```

```python
# Handle outliers for all numeric columns
numeric_cols = house.select_dtypes(include=['float64', 'int64']).columns

# Loop through each numeric column and clip outliers
for col in numeric_cols:
    Q1 = house[col].quantile(0.25)
    Q3 = house[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    house[col] = house[col].clip(lower=lower_bound, upper=upper_bound)

# Confirm no extreme outliers remaining
outliers_after = {}

for col in numeric_cols:
    Q1 = house[col].quantile(0.25)
    Q3 = house[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = house[(house[col] < lower_bound) | (house[col] > upper_bound)]
    outliers_after[col] = len(outliers)

# Print remaining outliers
print("Outliers Remaining After Clipping:")
for col, count in outliers_after.items():
    if count > 0:
        print(f"{col}: {count}")
```

#### Sales volume outliers were retained after careful review, as they reflect genuine market volatility rather than data errors. Preserving these values maintains the real-world integrity of the dataset, ensuring that subsequent analysis and models accurately capture true market behavior.

## ARIMA

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Prepare the Data
house_arima = house.copy()
house_arima.set_index('pivotable_date', inplace=True)
```

```python
# Target Variable
price_series = house_arima['average_price_all_property_types']
```

```python
# Log-transform the series
log_price_series = np.log(price_series)
```

```python
# Plot the Time Series
plt.figure(figsize=(10, 5))
plt.plot(price_series, marker='o', label='Average Price')
plt.title('Average House Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (£)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Check Stationarity after Log Transformation
adf_result = adfuller(log_price_series)
print(f"ADF Statistic after log: {adf_result[0]}")
print(f"p-value after log: {adf_result[1]}")
if adf_result[1] > 0.05:
    print("Still not stationary after log. Differencing needed.")
else:
    print("Stationary after log. No differencing needed.")
```

```python
# Difference the Log Series
log_price_diff = log_price_series.diff().dropna()
```

```python
# Check Stationarity again
adf_result_diff = adfuller(log_price_diff)
print(f"ADF Statistic after log and diff: {adf_result_diff[0]}")
print(f"p-value after log and diff: {adf_result_diff[1]}")
if adf_result_diff[1] > 0.05:
    print("Still not stationary after log + 1st diff. Might need second differencing.")
else:
    print("Stationary after log + 1st diff.")
```

```python
# Apply second differencing
log_price_diff2 = log_price_series.diff().diff().dropna()

# Check stationarity after second differencing
adf_result_diff2 = adfuller(log_price_diff2)
print(f"ADF Statistic after log and second diff: {adf_result_diff2[0]}")
print(f"p-value after log and second diff: {adf_result_diff2[1]}")

if adf_result_diff2[1] > 0.05:
    print("Still not stationary after second differencing")
else:
    print("Stationary after second differencing")
```
